```python
import re
import pandas as pd
import numpy as np
from collections import Counter
```

```python
def count_words(lines, delimiter=" "):
    
    words = Counter() # instantiate a Counter object called words
    for line in lines:
        for word in line.split(delimiter):
            words[word] += 1 # increment count for word
    return words
```

```python
df = pd.read_csv('../datasets/mcdonalds-yelp-negative-reviews.csv')
```

```python
word_count_crude = count_words(df['review'])
```

1. Identify 4 recurring themes/topics that reviewers are unhappy with. For example, one theme is that users are consistently unhappy with the drive-through experience.
- drive-through
- breakfast
- sauce

2. Next, using regex, group together all occurrences of these phrases. For example, `drive-thru`, `drive through`, `drivethrough` can all be replaced as `_DRIVE_THROUGH_`.

```python
# drive-thru
df['review_temp1'] = df['review'].str.replace(r'\bdrive(\s*|\-)(through|thru)\b', '_DRIVE_THROUGH_', case=False)
```

```python
# breakfast
df['review_temp2'] = df['review_temp1'].str.replace(r'\b(breakfast|eggs?|coffee|morning|(Mc\s*)?(Muffin|Griddle)|biscuits?)\b', '_MORNING_', case=False)
```

```python
# sauce
df['review_temp3'] = df['review_temp2'].str.replace(r'\b(sauces?|ketchup)\b', '_SAUCE_', case=False)
```

```python
# waiting time
df['review_temp4'] = df['review_temp3'].str.replace(r'\b(lines?|queu(e|ing|ed)|wait|minutes?|hours?)\b', '_WAIT_', case=False)
```

```python
df['final'] = df['review_temp4']
```

```python
df['city'].unique()
```

```python
sum(df['final'].str.count('_MORNING_'))
```

```python
words = ['_DRIVE_THROUGH_', '_MORNING_', '_SAUCE_', '_WAIT_']
cities = list(df['city'].unique())[:-1]
for word in words:
    print('%s: %s'%(word, sum(df['final'].str.count(word))))
for city in cities:
    for word in words:
        print('%s, %s: %s, %s'%(city, word, sum(df[df['city'] == city]['final'].str.count(word)), sum(df[df['city'] == city]['final'].str.count(word))/len(df[df['city'] == city])))
```
